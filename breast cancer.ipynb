```python
df["diagnosis"]=df["diagnosis"].map({'B':0,'M':1}).astype(int)
df.head()
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
X=df[['radius_mean','perimeter_mean','area_mean','compactness_mean','concave points_mean','radius_se','perimeter_se', 'area_se','compactness_se', 'concave points_se','radius_worst','perimeter_worst','compactness_worst','concave points_worst','texture_worst','area_worst','symmetry_worst','fractal_dimension_worst']]
y=df[['diagnosis']]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=22)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
reg= KNeighborsClassifier(n_neighbors=5)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
reg.fit(X_train,y_train)
```

```python
y_train_prediction=reg.predict(X_train)
```

```python
training_data=pd.DataFrame()
training_data=X_train
training_data["actual points"]=y_train
training_data["prediction_data"]=y_train_prediction
```

```python
training_data.head(5)
```

```python
training_data.index
```

```python
tp=0
tn=0
fp=0
fn=0
for i in training_data.index:
    if training_data['actual points'][i]==1 and training_data['prediction_data'][i]==1:
        tp+=1
    elif training_data['actual points'][i]==0 and training_data['prediction_data'][i]==0:
        tn+=1
    elif training_data['actual points'][i]==0 and training_data['prediction_data'][i]==1:
        fp+=1
    else:
        fn+=1
print(f'total true psitives:{tp}')
print(f'total true negitives:{tn}')
print(f'total false psitives:{fp}')
print(f'total false megitives:{fn}')
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(training_data['actual points'],training_data["prediction_data"])
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(training_data['actual points'],training_data["prediction_data"])
```
